```python
import json

employees = [

    {
        "employee_id": 101,
        "name": "Rahul Sharma",
        "department": "Data Engineering",
        "salary": 75000,
        "city": "Hyderabad"
    },

    {
        "employee_id": 102,
        "name": "Priya Reddy",
        "department": "AI Engineering",
        "salary": 85000,
        "city": "Bangalore"
    },

    {
        "employee_id": 103,
        "name": "Amit Kumar",
        "department": "Data Engineering",
        "salary": 65000,
        "city": "Mumbai"
    },

    {
        "employee_id": 104,
        "name": "Sneha Patel",
        "department": "Data Science",
        "salary": 95000,
        "city": "Chennai"
    },

    {
        "employee_id": 105,
        "name": "Farhan Ali",
        "department": "Cloud Engineering",
        "salary": 80000,
        "city": "Delhi"
    }

]

with open("employees.json", "w") as file:
    json.dump(employees, file, indent=4)

print("Employee data has been written to employees.json")
```

```python
import json

with open("employees.json", "r") as file:
    employees = json.load(file)
print(employees)
```

```python
for employee in employees:
    print(employee)
```

```python
for employee in employees:
    print(employee["name"])
```

```python
highest_salary = 0

for emp in employees:
    if emp["salary"] > highest_salary:
        highest_salary = emp["salary"]
print(highest_salary)
```

```python
#find employee with highest salary
for emp in employees:
    if emp["salary"] == highest_salary:
        print(emp)
```

```python
#find average salary
avg_salary = sum(emp["salary"] for emp in employees) / len(employees)
print(avg_salary)
```

```python
#find employees in Data Engineering department
for emp in employees:
    if emp["department"] == "Data Engineering":
        print(emp)
```

```python
#employees earning more than 80000
for emp in employees:
    if emp["salary"] > 80000:
        print(emp)
```

```python
#update a employee
for emp in employees:
    if emp["employee_id"] == 103:
        emp["salary"] = 70000
        break
print(employees[2])
```

```python
#add an employee
employees.append({
    "employee_id": 106,
    "name": "Neha Singh",
    "department": "AI Engineering",
    "salary": 90000,
    "city": "Bangalore"
})
print(employees[-1])
```

```python
#delete an employee
for employee in employees:
    if employee["employee_id"] == 102:
        employees.remove(employee)
        break
print(employees)
```
