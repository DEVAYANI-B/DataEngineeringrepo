```python
a = 10
b = 0
try:
    result = a/b
    print(result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
```

```python
#another ex

try:
    age = int(input("Enter your age: "))
    print(age)
except ValueError:
    print("Error: Please enter a valid integer for age.")
```

```python
#multiple exceptions
try:
    age = int(input("Enter your age: "))
    print(100/age)
except ValueError:
    print("Error: Please enter a valid integer for age.")
except ZeroDivisionError:
    print("Error: Age cannot be zero when calculating 100 divided by age.")

```

```python
#exception object
try:
    num = int("abc")
except Exception as e:
    print(e)

#else block
try:
    num = 10
    print(num)
except:
    print("error")
else:
    print("Success")

#finally
try:
    print(10/0)
except:
    print("Error")
finally:
    print("connection closed")

#raise error
salary = -5000

try:
    if salary < 0:
        raise ValueError("Salary cannot be negative.")
except ValueError as e:
    print(e)

```

FILE OPERATIONS


```python
file = open("employees.txt", "r")

print(file.read())
file.close()
```

```python

#right way
try:
    data = file.read()
    print(data)
except Exception as e:
    print(f"An error occurred: {e}")
finally:
    file.close()
```

```python
#read a single line
file = open("employees.txt", "r")

print(file.readline())
file.close()
```

```python
#read all lines
file = open("employees.txt", "r")
lines = file.readlines()
print(lines)
file.close()
```

```python
#automatically close the file using with statement
with open("employees.txt", "r") as file:
    data = file.read()
    print(data)


```

```python
#writing in file, executing this code again and again will only over write the same file
with open("new_employees.txt", "w") as file:
    file.write("Rahul\n")
    file.write("Priya\n")

```

```python
#append will not overwrite the file, adds new data at the end of the file
with open("new_employees.txt", "a") as file:
    file.write("Vishal\n")
```
