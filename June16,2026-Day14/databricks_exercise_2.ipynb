```python
#91
doctors_etl = spark.read.csv("/FileStore/datasets/doctors.csv", header=True, inferSchema=True)
visits_etl = spark.read.csv("/FileStore/datasets/visits.csv", header=True, inferSchema=True)
hospitals_etl = spark.read.option("multiline", "true").json("/FileStore/datasets/hospital_config.json")
print("ETL: All sources loaded")
```

```python
#92
visits_etl = visits_etl.fillna({"bill_amount": 0, "payment_status": "Unknown"})
hospitals_flat_etl = hospitals_etl.select(
    "hospital_id", "hospital_name", "city",
    coalesce(col("contact.phone"), lit("N/A")).alias("phone"),
    coalesce(col("contact.email"), lit("N/A")).alias("email"),
    "services"
)
print("ETL: Nulls cleaned")
```

```python
#93
hospitals_exploded_etl = hospitals_flat_etl.withColumn("service", explode("services")).drop("services")
hospitals_exploded_etl.show()
```

```python
#94
joined_etl = visits_etl.join(doctors_etl, "doctor_id", "inner")
joined_etl.show()
```

```python
#95
revenue_etl = joined_etl.withColumn("tax", round(col("bill_amount") * 0.05, 2)) \
    .withColumn("final_bill", col("bill_amount") + col("tax"))
revenue_etl.show()
```

```python
#96
doctor_rev_etl = revenue_etl.groupBy("doctor_id", "doctor_name", "specialization", "city") \
    .agg(sum("final_bill").alias("total_revenue"))
window_rank = Window.orderBy(desc("total_revenue"))
doctor_rev_etl = doctor_rev_etl.withColumn("revenue_rank", rank().over(window_rank))
doctor_rev_etl.show()
```

```python
#97
spec_summary = revenue_etl.groupBy("specialization") \
    .agg(
        count("visit_id").alias("total_visits"),
        sum("bill_amount").alias("total_revenue"),
        round(avg("bill_amount"), 2).alias("avg_revenue"),
        round(avg("consultation_fee"), 2).alias("avg_consultation_fee")
    )
spec_summary.show()
```

```python
#98
doctor_rev_etl.write.mode("overwrite").parquet("/FileStore/datasets/silver/doctor_revenue")
spec_summary.write.mode("overwrite").parquet("/FileStore/datasets/silver/specialization_summary")
hospitals_exploded_etl.write.mode("overwrite").parquet("/FileStore/datasets/silver/hospitals")
print("Silver layer saved!")
```

```python
#99
gold_top_doctors = doctor_rev_etl.filter(col("revenue_rank") <= 3)
gold_top_doctors.write.mode("overwrite").parquet("/FileStore/datasets/gold/top_doctors")

gold_spec_report = spec_summary.orderBy(desc("total_revenue"))
gold_spec_report.write.mode("overwrite").parquet("/FileStore/datasets/gold/specialization_report")
print("Gold reports saved!")
```

```python
#100
from pyspark.sql.functions import countDistinct

hospital_ops = doctor_rev_etl.join(
    hospitals_exploded_etl.groupBy("city").agg(countDistinct("hospital_id").alias("hospital_count")),
    "city", "left"
).join(
    spec_summary.select("specialization", "total_visits", "avg_revenue"),
    "specialization", "left"
).select(
    "doctor_id", "doctor_name", "specialization", "city",
    "total_revenue", "revenue_rank", "hospital_count",
    "total_visits", "avg_revenue"
).orderBy("revenue_rank")

hospital_ops.show(truncate=False)
hospital_ops.write.mode("overwrite").option("header", True).csv("/FileStore/datasets/gold/hospital_ops_dashboard")
print("Hospital Operations Analytics Dashboard saved!")
```
