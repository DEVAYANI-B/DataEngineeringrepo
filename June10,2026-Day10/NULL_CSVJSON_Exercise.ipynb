```python
quality_df = df.withColumn("data_quality_status", when(col("city").isNull() | col("service_type").isNull() | col("provider").isNull() | col("booking_amount").isNull() | col("booking_status").isNull() | col("payment_mode").isNull(), "Incomplete").otherwise("Complete"))

quality_df.show()
```

```python
quality_df.groupBy("data_quality_status").count().show()
```

```python
quality_df.filter(col("data_quality_status") == "Complete").show()
```

```python
quality_df.filter(col("data_quality_status") == "Incomplete").show()
```

```python
tax_df = df.fillna({"booking_amount": 0}).withColumn("tax", col("booking_amount") * 0.05)

tax_df.show()
```

```python
final_df = tax_df.withColumn("final_amount", col("booking_amount") + col("tax"))

final_df.show()
```

```python
df.filter(col("booking_status") == "Confirmed").agg(sum("booking_amount").alias("revenue")).show()
```

```python
avg1 = df.fillna({"booking_amount": 0}).agg(avg("booking_amount")).show()
```

```python
avg2 = df.dropna(subset=["booking_amount"]).agg(avg("booking_amount")).show()
```

```python
print("Average after replacing nulls with 0:")
df.fillna({"booking_amount": 0}).agg(avg("booking_amount").alias("average")).show()

print("Average after dropping null rows:")
df.dropna(subset=["booking_amount"]).agg(avg("booking_amount").alias("average")).show()
```
