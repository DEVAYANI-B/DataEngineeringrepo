```python
import pandas as pd

data = [
    (101, "Rahul Sharma", "Data Engineering", 75000, "Hyderabad"),
    (102, "Priya Reddy", "AI Engineering", 85000, "Bangalore"),
    (103, "Amit Kumar", "Data Engineering", 65000, "Mumbai"),
    (104, "Sneha Patel", "Data Science", 95000, "Chennai"),
    (105, "Farhan Ali", "Cloud Engineering", 80000, "Delhi")
]

columns = [
    "employee_id",
    "name",
    "department",
    "salary",
    "city"
]

df = pd.DataFrame(data, columns=columns)
display(df.head())
```

```python
df['bonus'] = df['salary'] * 0.10
display(df.head())
```

```python
df['tax'] = df['salary'] * 0.05
display(df.head())
```

```python
df['annual_salary'] = df['salary'] * 12
display(df.head())
```

```python
df['salary_after_bonus'] = df['salary'] + df['bonus']
display(df.head())
```

```python
df.rename(columns={'salary': 'monthly_salary'}, inplace=True)
display(df.head())
```

```python
df.rename(columns={'department': 'team'}, inplace=True)
display(df.head())
```

```python
total_salary_payout = df['monthly_salary'].sum()
print(f"Total Salary Payout: ₹{total_salary_payout}")
```

```python
average_salary = df['monthly_salary'].mean()
print(f"Average Salary: ₹{average_salary}")
```

```python
max_salary = df['monthly_salary'].max()
print(f"Maximum Salary: ₹{max_salary}")
```

```python
min_salary = df['monthly_salary'].min()
print(f"Minimum Salary: ₹{min_salary}")
```

```python
employee_count = df['employee_id'].count()
print(f"Total Employee Count: {employee_count}")
```

```python
employees_by_department = df['team'].value_counts()
print("Employees by Department:")
display(employees_by_department)
```

```python
avg_salary_by_department = df.groupby('team')['monthly_salary'].mean()
print("Average Salary by Department:")
display(avg_salary_by_department)
```

```python
max_salary_by_department = df.groupby('team')['monthly_salary'].max()
print("Maximum Salary by Department:")
display(max_salary_by_department)
```

```python
min_salary_by_department = df.groupby('team')['monthly_salary'].min()
print("Minimum Salary by Department:")
display(min_salary_by_department)
```

```python
total_salary_by_department = df.groupby('team')['monthly_salary'].sum()
print("Total Salary by Department:")
display(total_salary_by_department)
```

```python
unique_departments = df['team'].unique()
print("Unique Departments:")
print(unique_departments)
```

```python
unique_cities = df['city'].unique()
print("Unique Cities:")
print(unique_cities)
```

```python
count_unique_departments = df['team'].nunique()
print(f"Count of Unique Departments: {count_unique_departments}")
```

```python
top_3_highest_paid = df.nlargest(3, 'monthly_salary')
print("Top 3 Highest-Paid Employees:")
display(top_3_highest_paid)
```

```python
lowest_paid_employee = df.nsmallest(1, 'monthly_salary')
print("Lowest-Paid Employee:")
display(lowest_paid_employee)
```

```python
employees_start_with_R = df[df['name'].str.startswith('R')]
print("Employees whose names start with 'R':")
display(employees_start_with_R)
```

```python
employees_contain_a = df[df['name'].str.contains('a', case=False)]
print("Employees whose names contain 'a':")
display(employees_contain_a)
```

```python
employees_in_engineering = df[df['team'].str.contains('Engineering', case=False, na=False)]
print("Employees working in departments containing 'Engineering':")
display(employees_in_engineering)
```

```python
def categorize_salary(salary):
    if salary >= 85000:
        return 'High'
    elif salary >= 75000:
        return 'Medium'
    else:
        return 'Low'

df['salary_category'] = df['monthly_salary'].apply(categorize_salary)
display(df.head())
```

```python
salary_category_counts = df['salary_category'].value_counts()
print("Count of Employees in Each Salary Category:")
display(salary_category_counts)
```

```python
final_report = df[['employee_id', 'name', 'team', 'monthly_salary', 'bonus', 'annual_salary']]
print("Final Report:")
display(final_report)
```
