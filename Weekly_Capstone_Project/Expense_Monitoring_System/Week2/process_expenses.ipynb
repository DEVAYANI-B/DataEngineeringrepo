```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('expenses.csv')
```

```python
df['amount'] = df['amount'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.to_period('M')
```

```python
print("Cleaned data:")
print(df.head(15))
```

```python
print("\nMonthly totals per category")
monthly_summary = df.groupby(['month', 'category'])['amount'].sum().unstack().fillna(0)
print(monthly_summary)
```

```python
amounts = df['amount'].to_numpy()
print("\nOverall stats (numpy)")
print("Total spent:    ", round(np.sum(amounts), 2))
print("Average expense:", round(np.mean(amounts), 2))
print("Highest expense:", round(np.max(amounts), 2))
print("Lowest expense: ", round(np.min(amounts), 2))
print("Std deviation:  ", round(np.std(amounts), 2))
```

```python
monthly_summary.to_csv('monthly_summary.csv')
print("\nSaved monthly_summary.csv")
```
