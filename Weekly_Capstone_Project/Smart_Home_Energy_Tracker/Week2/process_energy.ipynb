```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('energy_usage.csv')

print("Raw rows:", len(df))
print(df.isna().sum())
```

```python
df = df.dropna(subset=['energy_kwh'])
df['energy_kwh'] = df['energy_kwh'].astype(float)
df['timestamp'] = pd.to_datetime(df['timestamp'])

print("\nRows after cleaning:", len(df))
```

```python
print("\nPer-device totals and averages (numpy)")
for device_id, group in df.groupby('device_id'):
    readings = group['energy_kwh'].to_numpy()
    print(f"Device {device_id}: total={np.sum(readings):.2f} kWh, "
          f"avg={np.mean(readings):.2f} kWh, max={np.max(readings):.2f} kWh")
```

```python
print("\nRoom-level summary (pandas)")
room_summary = df.groupby('room_id')['energy_kwh'].sum().sort_values(ascending=False)
print(room_summary)
```

```python
room_summary.to_csv('room_summary.csv', header=['total_kwh'])
print("\nSaved room_summary.csv")
```
