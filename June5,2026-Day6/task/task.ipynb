18th

```python
dept_count = {}
for emp in employees:
    dept = emp[2]
    dept_count[dept] = dept_count.get(dept, 0) + 1

for dept, count in dept_count.items():
    print(dept, "=", count)
```

```python
highest_salary = max(int(emp[3]) for emp in employees)
for emp in employees:
    if int(emp[3]) == highest_salary:
        print(emp)
```
